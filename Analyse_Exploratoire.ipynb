### Question 2 : Quelle est la répartition des années de sortie des films disponibles sur TMDB par rapport aux films ajoutés sur Netflix ? Représenter graphiquement 

On exclut les films de TMDB dont :
-la date est inférieure à 1925 (pour avoir une cohérence avec la base de films de Netflix)

-la date dépasse 2024 (car il y'avait certains films qui étaient seulement annoncés ou des erreurs de renseignement)

-la date est égale à 0 car on a attribué l'année 0 pour les films dont l'année n'était pas renseignée.

```python
df_tmdb_freq = df_tmdb.loc[(df_tmdb["release_year"]!=0) & 
                       (df_tmdb["release_date"]>= "1925") & 
                       (df_tmdb["release_date"]<"2024")]
```

Ensuite, on compte le nombre de films sortis par année sur chaque plateforme, et on les trie par ordre croissant :

```python
df_tmdb_fpa = df_tmdb_freq["release_year"].value_counts().sort_index()
df_netflix_fpa = df_netflix["release_year"].value_counts().sort_index()
```

```python
#Création graphique de la "fonction" de répartition

x1 = np.linspace(1925, 2024, len(df_tmdb_fpa))
x2 = np.linspace(1925, 2024, len(df_netflix_fpa))
fig, axes = plt.subplots(2, 1, figsize=(12, 6))

axes[0].plot(x1, df_tmdb_fpa.values, color="blue", linewidth="2",label="TMDB")
axes[1].plot(x2, df_netflix_fpa.values, color="red", linewidth="2",label="Netflix")

axes[0].legend()
axes[1].legend()

plt.xlabel("Année")
plt.ylabel("Nombre de films")

plt.show()
```

On constate que Netflix a tendance à favoriser les films récents, le nombre de films disponibles sur Netflix dont l'année de sortie est comprise entre 2000 et 2020 (notamment aux alentours de 2015), est beaucoup plus élevé que sur les décénnies précédentes. La chute après 2020 s'explique probablement par des raisons légales d'acquisition après la sortie au cinéma.

La tendance croissante pour les films de TMDB est simplement due à l'expansion du cinéma, et donc la hausse continue de films produits.

### Question 3 -Analyse des facteurs influant sur la note, Analyse des facteurs inuant sur la popularité : Quels critères influent sur la note d’un film ? Utiliser des graphiques pour illustrer ces relations, et mettre en avant les tendances importantes identifiées 

Pour cette question, nous avons fait des nuages de points qui presentent un à un les liens qui existent entre la note/popularité d'un film, et d'autres variables.

#### Variable Adulte

```python
df_tmdb_va = df_tmdb.groupby("adult").agg({"vote_average": "mean"})

df_tmdb_p = df_tmdb.groupby("adult").agg({"popularity": "mean"})
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].bar(df_tmdb_va.index.astype(str), df_tmdb_va["vote_average"])
axes[0].set_title("Note moyenne par catégorie adulte")
axes[0].set_xlabel("Adulte")
axes[0].set_ylabel("Note moyenne")

axes[1].bar(df_tmdb_p.index.astype(str), df_tmdb_p["popularity"])
axes[1].set_title("Popularité moyenne par catégorie adulte")
axes[1].set_xlabel("Adulte")
axes[1].set_ylabel("Popularité moyenne")
plt.show()
```

Ici, on observe que la popularité d'un film est bien plus élevée quand il n'est pas réservé à un public adulte, ce qui est logique car il touche forcément plus de monde. Cependant, cette restriction donne une note moyenne de films plus élevée, ce qui s'explique peut être par le fait que le nombre de votes a tendance à tirer la moyenne vers le bas.

##### Durée du film

```python
df_tmdb_r = df_tmdb.loc[(df_tmdb["runtime"].notna()) #Afin de retirer les films dont les durées ne sont pas renseignées.
                        & (df_tmdb["runtime"]>0) 
                        & (df_tmdb["runtime"]<=240)] #Pour ne pas prendre les valeurs aberrantes

df_tmdb_rva = df_tmdb_r.groupby("runtime").agg({"vote_average":"mean"})
df_tmdb_rpa = df_tmdb_r.groupby("runtime").agg({"popularity":"mean"})


```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].scatter(df_tmdb_rva.index, df_tmdb_rva.values)
axes[0].set_title("Note moyenne par durée de film")
axes[0].set_xlabel("Runtime")
axes[0].set_ylabel("Note moyenne")

axes[1].scatter(df_tmdb_rpa.index, df_tmdb_rpa.values)
axes[1].set_title("Popularité moyenne par durée de film")
axes[1].set_xlabel("Runtime")
axes[1].set_ylabel("Popularité moyenne")
plt.show()
```

Les notes moyennes des films augmentent avec leur durée, ce qui peut suggérer que les spectateurs préfèrent généralement les films plus longs en termes de qualité (peut-être car les histoires sont mieux développées).

La popularité moyenne suit une tendance différente : elle atteint un maximum pour des films d’environ 100 à 120 minutes, ce qui pourrait indiquer que cette durée correspond à un format idéal pour attirer un public large. Les films très longs perdent en popularité, potentiellement à cause du public qui fatigue ou se lasse.

##### Budget

```python
df_tmdb_b = df_tmdb.loc[(df_tmdb["budget"].notna())
                        & (df_tmdb["budget"]>0)
                        & (df_tmdb["budget"]<=50000000)]


df_tmdb_ppb = df_tmdb_b.loc[(df_tmdb_b["popularity"]>0) 
                            & (df_tmdb_b["popularity"]<100)] #On retire les valeurs aberrantes

df_tmdb_bva = df_tmdb_b.groupby("budget").agg({"vote_average":"mean"})
df_tmdb_bpa = df_tmdb_ppb.groupby("budget").agg({"popularity":"mean"})
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].scatter(df_tmdb_bva.index, df_tmdb_bva.values)
axes[0].set_title("Note moyenne en fonction du budget")
axes[0].set_xlabel("Budget")
axes[0].set_ylabel("Note moyenne")

axes[1].scatter(df_tmdb_bpa.index, df_tmdb_bpa.values)
axes[1].set_title("Popularité moyenne en fonction du budget")
axes[1].set_xlabel("Budget")
axes[1].set_ylabel("Popularité moyenne")
plt.show()
```

Le nuage de points ne nous dit pas grand chose...

##### Langue originale du film

```python
df_tmdb_l = df_tmdb.loc[df_tmdb["original_language"].isin(["en", "fr", "es", "ko", "ja", "it", "de", "zh", "pt", "ar", "hi"])]

df_tmdb_val = df_tmdb_l.groupby("original_language").agg({"vote_average": "mean"})

df_tmdb_pl = df_tmdb_l.groupby("original_language").agg({"popularity": "mean"})
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_tmdb_val.index, df_tmdb_val["vote_average"])
plt.title("Note moyenne par langue originale")
plt.xlabel("Langue originale")
plt.ylabel("Note moyenne")
plt.show()

plt.figure(figsize=(12,6))
plt.bar(df_tmdb_pl.index, df_tmdb_pl["popularity"])
plt.title("Popularité moyenne par langue originale")
plt.xlabel("Langue originale")
plt.ylabel("Popularité moyenne")
plt.show()
```

Les notes moyennes ne sont pas vraiment affectées par la langue originale, sûrement car cela a peu d'intérêt pour comprendre le film étant donné qu'on peut changer la langue.
Par contre, la langue originale joue un énorme rôle dans la popularité des films, sûrement à cause de leur réputation et l'intérêt qu'y porte le public. On peut notamment citer l'avènement des "K-Drama" qui explique sûrement en grande partie le top 1 du coréen. Le top 3 est complété par le japonais et l'anglais.
